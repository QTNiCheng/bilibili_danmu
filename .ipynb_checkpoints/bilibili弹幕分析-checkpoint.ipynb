```python
import pandas as pd
import pyecharts as pye
```

```python
data = pd.read_csv(r'data/【哔哩哔哩2015拜年祭】1P.csv', encoding='GBK')
data.head()
```

```python
data.insert(5, '弹幕发送日期', data['发送时间'].str[:10]) 
data.insert(6, '弹幕发送时间', data['发送时间'].str[10:16])
data.head()
```

```python
font_size = data.groupby(by=['字号']).size()
danmu_color = data.groupby(by=['颜色']).size()
send_date = data.groupby(by=['弹幕发送日期']).size()
send_time = data.groupby(by=['弹幕发送时间']).size()
sender_id = data.groupby(by=['发送者id']).size()
```

```python
line = pye.Line('发送弹幕的日期')
line.add('', send_date.index, send_date.values, xaxis_interval=30, xaxis_rotate=30, mark_point=['max'])
# line.add('时间', send_time.index, send_time.values)
line
```

# 各时间段弹幕数量
- 可以看出在2016年1月到2月的时候弹幕数量是最多的，那会刚好是春节，可能大家都在等16年的拜年祭无聊而回顾15年的拜年祭吧。

```python
line = pye.Line('发送弹幕的时间')
line.add('', send_time.index, send_time.values, xaxis_interval=60, xaxis_rotate=30, mark_point=['max'])
line
```

# 一天内弹幕的数量分布
- 可以看出12点到24点的人数是比较多的，也就是说这段时间用户活跃度高。毕竟熬夜的人还是少部分的，早上喜欢睡懒觉的人还是多的
