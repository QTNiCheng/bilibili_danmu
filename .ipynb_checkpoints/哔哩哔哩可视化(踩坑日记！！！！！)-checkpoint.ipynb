```python
import pandas as pd
import pyecharts as pye
```

```python
data = pd.read_csv(r'data\【哔哩哔哩2015拜年祭】1P.csv', encoding='GBK')
data.head(5)
```

```python
data.info()
```

```python
time_list = [data['发送时间'][a][:10] for a in range(0, len(data['发送时间']))]
time_list
```

```python
len(time_list)
```

```python
a = {}
for i in time_list:
    if time_list.count(i)>1:
        a[i] = time_list.count(i)
print(a)
```

```python
time_data = pd.DataFrame(time_list)
time_data
```

```python
time_data.groupby(0).sum()
```

```python
gp = time_data.groupby(by=[0])
time_count = gp.size()
print(type(time_count))
time_count
```

```python
time_count.index
```

```python
bar = pye.Bar('历史弹幕数量')
bar.add('', time_count.index, time_count.values, xaxis_interval=30, xaxis_rotate=30, is_datazoom_show=True)
bar
```
